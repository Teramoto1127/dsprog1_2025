# BeautifulSoupでHTMLソースから任意の情報を取得する

```python
url = "https://www.musashino-u.ac.jp/" 
res = requests.get(url, headers=headers)
res.encoding = res.apparent_encoding
print(f"ステータスコード: {res.status_code}")
#ステータスコードを表示
```

```python
# HTMLソースを解析
soup = BeautifulSoup(res.text, 'html.parser')
print(f"soupの型: {type(soup)}")
```

```python
#ソースコード全体から<a>タグを抽出
soup_a = soup.a
print(f"<a>タグ: {soup_a}")
print(f"<a>タグの型: {type(soup_a)}")  

#ソースコード全体から<p>タグを抽出
soup_p = soup.p
print(f"<p>タグ: {soup_p}")
print(f"<p>タグの型: {type(soup_p)}")
# ソースコード全体から<div>タグを抽出
soup_div = soup.div
print(f"<div>タグ: {soup_div}")
print(f"<div>タグの型: {type(soup_div)}")
```

```python
#ソースコード全体から<div> タグを抽出
soup_div = soup.div
#print(f"<div>タグ: {soup_div}")
# print(f"<div>タグの型: {type(soup_div)}")   

soup_div_div = soup_div.div
print(f"<div>タグの中の<div>タグ: {soup_div_div}")
print(f"<div>タグの中の<div>タグの型: {type(soup_div_div)}")
```

## 指定した属性で検索

```python
tag_href_list = soup.find_all(attres={'href':':'},limit=5)
for tag in tag_href_list:
    print(f"[tag]") #href属性が'/'のタグを表示
```

```python
#idやclass属性でも検索可能
#例えば, id属性が"main"の要素やclass属性が"header"の要素を取得する場合
tag_id_list = soup.find_all(id='top')
for tag in tag_id_list:
    print(f"{tag}")
tag_class_list = soup.find_all(class_='p-notices',limit=5)
for tag in tag_class_list:
    print(f"{tag}") #class属性が"p-notice"の要素
```

```python
# id属性を指定
id_tags = soup.select('#top')
for tag in id_tags:
    print(f"{tag}")


# class属性を指定
class_tags = soup.select('.p-notices')
for tag in class_tags:
    print(f"{tag}")


# HTML階層構造を指定
#例えば、<div>タグの中の<h2>タグを取得する場合
div_h2_tags = soup.select('div h2')
for tag in div_h2_tags:
    print(f"{tag}")

# ブラウザから取得したセレクタを使う
# 例えば、ブラウザの検証ツールで取得したが以下の場合
#body
```

```python
# movie_titles_scrape.py
import requests
from bs4 import BeautifulSoup
import re

URL = "https://abehiroshi.la.coocan.jp/movie/eiga.htm"

def fetch_text(url, timeout=10):
    resp = requests.get(url, timeout=timeout, headers={"User-Agent": "Mozilla/5.0"})
    # 自動で文字コード判定してセット（日本語サイトで有効）
    resp.encoding = resp.apparent_encoding
    return resp.text

def clean(s):
    # 前後空白除去、複数空白を1つに、全角/半角混在に配慮
    s = re.sub(r'\s+', ' ', s.strip())
    # 余分な記号があれば削る（必要に応じて調整）
    s = re.sub(r'[\u200b\u200e\u200f]', '', s)
    return s

def extract_titles(html):
    soup = BeautifulSoup(html, "html.parser")

    candidates = []

    # 1) 見出しタグから
    for h in soup.find_all(re.compile('^h[1-6]$')):
        text = clean(h.get_text())
        if text:
            candidates.append(text)

    # 2) liタグ（リスト）から
    for li in soup.find_all("li"):
        text = clean(li.get_text())
        if text:
            candidates.append(text)

    # 3) aタグ（リンク）から — リンクテキストのみ抽出
    for a in soup.find_all("a"):
        text = clean(a.get_text())
        if text:
            candidates.append(text)

    # 4) ページ本文の段落もチェック（必要なら）
    for p in soup.find_all("p"):
        text = clean(p.get_text())
        if text and len(text) < 120:  # あまり長い行は除外（タイトルは短めの想定）
            candidates.append(text)

    # 簡単なフィルタリング：
    # - 行頭に「・」や番号があれば除去
    normalized = [re.sub(r'^[\s\d\.\-・]+\s*', '', c) for c in candidates]
    # - 重複除去（順序保持）
    seen = set()
    titles = []
    for t in normalized:
        if not t:
            continue
        # 英数字だけの短すぎる文字列は除外（必要なら調整）
        if len(t) <= 1:
            continue
        if t in seen:
            continue
        seen.add(t)
        titles.append(t)

    return titles

def main():
    html = fetch_text(URL)
    titles = extract_titles(html)
    # 出力：1行に1タイトル
    for t in titles:
        print(t)

if __name__ == "__main__":
    main()
```
